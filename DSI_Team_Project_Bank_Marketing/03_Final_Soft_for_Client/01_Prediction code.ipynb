```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import joblib

# Load trained model
model_filename = 'logreg_model.pkl'  # Path to the saved model
logreg = joblib.load(model_filename)

# Load  new dataset
new_df = pd.read_csv('02_Template_Dataset_for_y_prediction.csv', sep=';')

# Process new dataset
X_new = new_df.drop('y', axis=1)
X_encoded_new = X_new.apply(LabelEncoder().fit_transform)
X_scaled_new = StandardScaler().fit_transform(X_encoded_new)

# Predict and add 'y' for new dataset
y_pred_new = logreg.predict(X_scaled_new)
new_df['y'] = y_pred_new
new_df.to_csv('03_Predicted _yes_no_clients.csv', index=False)

print("Predictions added and new dataset saved.")
df = pd.read_csv('03_Predicted _yes_no_clients.csv',sep=',')
y_counts = df['y'].value_counts()
print("Count of 'y' (yes):", y_counts.get('yes', 0))
print("Count of 'no':", y_counts.get('no', 0))
df
```
